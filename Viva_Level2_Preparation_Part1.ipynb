```python
from sklearn.compose import ColumnTransformer
```

```python
pr.columns
```

```python
impute_tf=[
    ('mode_imputer',SimpleImputer(strategy='most_frequent'),['postal_code','preferred_category','subscription_type','payment_method','country','gender']),
    ('median_imputer',SimpleImputer(strategy='median'),['avg_session_duration','age']),
    ('cp_imputer',SimpleImputer(strategy='constant',fill_value='missing_resonse'),['campaign_response']),
    ('pc_imputer',SimpleImputer(strategy='constant',fill_value=0),['purchase_count'])
]
impute_ct=ColumnTransformer(transformers=impute_tf,remainder='passthrough').set_output(transform='pandas')
impute_ct
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler, MinMaxScaler, RobustScaler
```

```python
pr=pr.drop(['customer_id','email','name','city','postal_code'],axis=1, errors='ignore')
```

```python
encode_tf=[
    ('ohe',OneHotEncoder(sparse_output=False),['country','subscription_type','payment_method','preferred_category','device_type']),
    ('oe',OrdinalEncoder(),['gender','is_active','campaign_response'])
]
encode_ct=ColumnTransformer(transformers=encode_tf,remainder='drop',verbose_feature_names_out=False).set_output(transform='pandas')
pipe1=Pipeline(steps=[('encoding',encode_ct),
                      ('scaling',StandardScaler())]).set_output(transform='pandas')
```

```python
pipe1
```

```python
from sklearn.preprocessing import StandardScaler
scale_tf=[('ss',StandardScaler(),['age','total_spent','avg_session_duration']),
          ('minmax',MinMaxScaler(),['purchase_count']),
          ('robust',RobustScaler(),['credit_score'])
          ]
scale_ct=ColumnTransformer(transformers=scale_tf,remainder='drop',verbose_feature_names_out=False).set_output(transform='pandas')
```

```python
scale_ct
```

```python
pipe2=Pipeline(steps=[('scaling',scale_ct)]).set_output(transform='pandas')
```

```python
pipe2
```

```python
pipe1.fit_transform(pr)
```

```python
pipe2.fit_transform(pr)
```

```python
from sklearn.pipeline import FeatureUnion
```

```python
fe=FeatureUnion(transformer_list=[('encode_tf',pipe1),
                                  ('scale_tf',pipe2)]).set_output(transform='pandas')
```

```python
fe
```

```python
fe.fit_transform(pr)
```
