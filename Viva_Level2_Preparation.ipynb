# PCA principle component features

```python
# principle features,  like suppose we have 4 features, so we can get 2 features,
# get best features
# it is also called dimensionality reduction
# X1, X2, X3, X4, so we cannot able to visualize all the 4 feauteres.
# X1, X2, X3-----------> X1,X4
# it can take a combination of the data and do the dimensionality redution.....
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
?pca
```

```python
from sklearn.datasets import load_iris
iris=load_iris(as_frame=True)
dtf7=iris.frame
dtf7.head()
```

```python
iris_df=dtf7.drop('target',axis=1)
```

```python
iris_df
```

```python
iris_pca=pca.fit_transform(iris_df)
```

```python
iris_pca
```
