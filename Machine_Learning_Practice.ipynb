```python
#Lets do a intersting code to understand the boolean indexing of an image of the Mandelbrot set
import numpy as np
import matplotlib.pyplot as plt
def mandelbrot(h,w,maxit=20,r=2):
  x=np.linspace(-2.5,1.5,4*h+1)
  y=np.linspace(-1.5,1.5,3*w+1)
  A,B=np.meshgrid(x,y)
  C=A+B*1j
  z=np.zeros_like(C)
  divtime=maxit+np.zeros(z.shape,dtype=int)
  for i in range(maxit):
    z=z**2+C
    diverge=abs(z)>r
    div_now=diverge & (divtime==maxit)
    divtime[div_now]=i
    z[diverge]=r
  return divtime
plt.clf()
plt.imshow(mandelbrot(400, 400))
```

## **Combining the DataFrame**

```python
import numpy as np
df={'Subject_id':[1,2,3,4,5],
    'First_Name':['Sarah','Reet','Robert','Shahida','Jia'],
    'Last_Name':['kanpur','Kaur','Dauglas','Ali','Gopalan']}
dtf=pd.DataFrame(df)
print(dtf)
print('\n')
sd={'Subject_id':[4,5,6,7,8],
    'First_Name':['Aani','Asma','Billy','Rudy','Prabhnoor'],
    'Last_Name':['Sinha','Anshari','John','Monga','Singh']}
sdf=pd.DataFrame(sd)
print(sdf)
print('\n')
td={'Subject_Id':[1,2,3,4,5,7,8,9,10,11],'Test_ID':[51,15,15,61,16,14,15,1,61,16]}
tdf=pd.DataFrame(td)
print(tdf)

print('\n')
sd3=pd.concat([dtf,sdf],axis=1,ignore_index=True)
print(sd3,'\n')
sd4=pd.concat([dtf,tdf],axis=1)
print(sd4)
```
