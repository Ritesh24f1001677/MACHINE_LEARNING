
# Mini Batch gradient descent

```python
import numpy as np
from sklearn.linear_model import SGDClassifier
batch_size=1
model =SGDClassifier()

# Determine all unique classes from the entire y_train dataset
# This ensures the 'classes' parameter in partial_fit is comprehensive.
all_classes = np.unique(y_train)
for i in range(0,X_train.shape[0],batch_size):
  X_batch=X_train[i:i+batch_size]
  y_batch=y_train[i:i+batch_size] # Corrected this line
  # Pass the `all_classes` to partial_fit
  model.partial_fit(X_batch,y_batch,classes=all_classes)
print('training score',model.score(X_train,y_train))
```

```python
import numpy as np
from sklearn.linear_model import SGDClassifier
batch_size=1
model =SGDClassifier(loss="log_loss")

# Determine all unique classes from the entire y_train dataset
# This ensures the 'classes' parameter in partial_fit is comprehensive.
all_classes = np.unique(y_train)
for i in range(0,X_train.shape[0],batch_size):
  X_batch=X_train[i:i+batch_size]
  y_batch=y_train[i:i+batch_size] # Corrected this line
  # Pass the `all_classes` to partial_fit
  model.partial_fit(X_batch,y_batch,classes=all_classes)
print('training score',model.score(X_train,y_train))
```

```python
fig,ax=plt.subplots()
ax.plot(ccp_alphas[:-1],impurities[:-1],marker="o",drawstyle="steps-post")
ax.set_xlabel('effective alpha')
ax.set_ylabel('total impurity of leaves')
ax.set_title('Total impurities Vs effective alpha for training set')
```

```python
models=[]
for ccp_alpha in ccp_alphas:
  model=DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
  model.fit(X_train,y_train)
  models.append(model)
print(
    'Number of the nodes in the last tree is: {} with ccp_alpha: {}'.format(
    models[-1].tree_.node_count, ccp_alphas[-1])
)
```

```python

train_scores=[model.score(X_train,y_train) for model in models]
test_scores=[model.score(X_test,y_test) for model in models]
fig,ax=plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy Vs Aplha for trainiing and testing sets")
ax.plot(ccp_alphas,train_scores,marker='o', label ="train")
ax.plot(ccp_alphas, test_scores,marker='o',label="test")
```
